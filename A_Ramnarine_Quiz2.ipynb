###  Consider the number-of-doors and price variables: Based on your EDA work, how many missing data values occur within each of these attributes?

```python
temp = df['num of doors'].replace('?', np.nan)
a = temp.dropna()
temp2 = df['price'].replace('?', np.nan)
b = temp2.dropna()
print('Missing data values in Num of Doors: ', df['num of doors'].count() - a.count())
print('Missing data values in Price: ', df['price'].count() - b.count())
```

### Replacing missing data values with mean

```python
df = df.replace('?', np.nan)
df['num of doors'].fillna('two', inplace = True)

#Was getting a type mismatch due to the NaN's
#So I converted all NaN's to 0, then changed the data type, then converted them back to NaN
df['price'].fillna(0, inplace = True)
df = df.astype({'price':'int'})
df = df.replace(0, np.nan)

df['price'].fillna(df['price'].mean(), inplace = True)
print(df)
```

### Scaling Data

Looking at the data set we can see that the values of engine size could benefit from scaling. This is because it's values are too high and therefore would not mend well when compared to other engine values such as bore or stroke. We want to scale them down to a reasonable level that wont be too low but also not too high. The number we arrive at still has to be proportional to the other engine values. So the best case of action would be to scale the values /10. We could implore the use of a MinMax or other type of scaler, however I believe it to be easier to just increase the values by hand.

First however let's plot these 2 attributes:

```python
plt.hist(df['engine size'])
```

```python
plt.boxplot(df['engine size'])
```

```python
df['stroke'].fillna(0, inplace = True)
df = df.astype({'stroke':'float'})
df = df.replace(0, np.nan)

plt.hist(df['stroke'])
```

```python
plt.boxplot(df['stroke'])
```

Now lets scale the data accordingly and see how it affects the graphs

```python
df = df.astype({'engine size':'float'})
df['engine size'] = df['engine size'] / 10

print(df['engine size'])
print(df)
```

```python
#remove all horsepower values that are not less than 2000
df['horsepower'].fillna(0, inplace = True)
df = df.astype({'horsepower':'int'})
df = df.replace(0, np.nan)
df = df[df['horsepower'] < 2000]  

#remove all rmp values greater than 1000
df['peak rpm'].fillna(0, inplace = True)
df = df.astype({'peak rpm':'int'})
df = df.replace(0, np.nan)
df = df[df['peak rpm'] > 1000]  

#remove all price values greater than 100,000
df = df[df['price'] < 100000]  
df
```

Consider the wheel-base, length, width, height, curb-weight, engine-size, compression ratio, horsepower, peak-rpm, and city-mpg variables. Using your dimensionality reduction expertise, use Python to reduce the dimensionality of this group of variables to a set of new orthogonal features.

Analyzing the data set you can see that certain attributes could be combined for dimensional reduction. For example, length width and height are all aspects of the same car engine. However we have a variable already labeled as engine size, so the exact dimensions of the engine are redundant. We can either get rid of the data completely, or reduce it down to a more reasonable size. We still want to keep a remenance of the data so reducing it to a smaller factor is the adequate solution. Another example can be peak rpm. Most rpm's fall within the range of 4-5k. So while it is beneficial in some aspects to have the exact rpm, reducing it to a general thousand scale would be more adequate. These are the 2 changes I will implement here.

```python
df = df.astype({'length':'float'})
df['length'] = df['length'] / 10

df = df.astype({'width':'float'})
df['width'] = df['width'] / 10

df = df.astype({'height':'float'})
df['height'] = df['height'] / 10

df['peak rpm'] = df['peak rpm'] / 1000

df
```
